```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from numpy._core.defchararray import upper

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
Direccion_guardado = "/workspaces/Marcos-J-Gomes_https-github.com-4GeeksAcademy-machine-learning-python_LinealRegression/data/processed"


X_train_sel_CON_outliers= pd.read_csv(f"{Direccion_guardado}/X_train_sel_CON_outliers_clean.csv")
X_test_sel_CON_outliers= pd.read_csv(f"{Direccion_guardado}/X_test_sel_CON_outliers_clean.csv")
X_train_sel_CON_outliers_norm=pd.read_csv(f"{Direccion_guardado}/X_train_sel_CON_outliers_clean.csv")
X_test_sel_CON_outliers_norm=pd.read_csv(f"{Direccion_guardado}/X_test_sel_CON_outliers_clean.csv")
X_train_sel_CON_outliers_scal=pd.read_csv(f"{Direccion_guardado}/X_train_sel_CON_outliers_scal_clean.csv")
X_test_sel_CON_outliers_scal=pd.read_csv(f"{Direccion_guardado}/X_test_sel_CON_outliers_scal_clean.csv")
X_train_sel_SIN_outliers=pd.read_csv(f"{Direccion_guardado}/X_train_sel_SIN_outliers_clean.csv")
X_test_sel_SIN_outliers=pd.read_csv(f"{Direccion_guardado}/X_test_sel_SIN_outliers_clean.csv")
X_train_sel_SIN_outliers_norm=pd.read_csv(f"{Direccion_guardado}/X_train_sel_SIN_outliers_norm_clean.csv")
X_test_sel_SIN_outliers_norm=pd.read_csv(f"{Direccion_guardado}/X_test_sel_SIN_outliers_norm_clean.csv")
X_train_sel_SIN_outliers_scal=pd.read_csv(f"{Direccion_guardado}/X_train_sel_SIN_outliers_scal_clean.csv")
X_test_sel_SIN_outliers_scal=pd.read_csv(f"{Direccion_guardado}/X_test_sel_SIN_outliers_scal_clean.csv")
y_train = pd.read_excel(f"{Direccion_guardado}/y_train.xlsx")
y_test  = pd.read_excel(f"{Direccion_guardado}/y_test.xlsx")

target= "charges"
```

# Regresion Lineal

normal

```python
datasets = {
    "CON_outliers": (X_train_sel_CON_outliers.drop(target, axis = 1), X_test_sel_CON_outliers.drop(target, axis = 1)),
    "SIN_outliers": (X_train_sel_SIN_outliers.drop(target, axis = 1), X_test_sel_SIN_outliers.drop(target, axis = 1))
}

Resultados = {}

for dataset_, (X_train, X_test) in datasets.items():
    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    Resultados[dataset_] = {
        "Model" : model,
        "R2": r2_score(y_test, y_pred),
        "MAE": mean_absolute_error(y_test, y_pred),
        "RMSE": np.sqrt(mean_squared_error(y_test, y_pred)),
        
    }

mejor_dataset = max(Resultados, key=lambda x: Resultados[x]["R2"])
mejor_r2 = Resultados[mejor_dataset]["R2"]
mejor_modelo = Resultados[mejor_dataset]["Model"]

print("Resultados por dataset:")
for k, v in Resultados.items():
    print(f"{k:25s} -> R2: {v['R2']:.5f} | MAE: {v['MAE']:.4f} | RMSE: {v['RMSE']:.4f}")

print("\nMejor dataset:", mejor_dataset)
print("Mejor R2:", mejor_r2)
print("\nParámetros del mejor modelo:")
print("Intercepto (a):", mejor_modelo.intercept_)
print("Coeficientes (b):",mejor_modelo.coef_)
```

Ridge

```python
datasets_ = {
    "CON_outliers": (X_train_sel_CON_outliers.drop(target, axis = 1), X_test_sel_CON_outliers.drop(target, axis = 1)),
    "CON_outliers_norm": (X_train_sel_CON_outliers_norm.drop(target, axis = 1), X_test_sel_CON_outliers_norm.drop(target, axis = 1)),
    "CON_outliers_scal": (X_train_sel_CON_outliers_scal.drop(target, axis = 1), X_test_sel_CON_outliers_scal.drop(target, axis = 1)),

    "SIN_outliers": (X_train_sel_SIN_outliers.drop(target, axis = 1), X_test_sel_SIN_outliers.drop(target, axis = 1)),
    "SIN_outliers_norm": (X_train_sel_SIN_outliers_norm.drop(target, axis = 1), X_test_sel_SIN_outliers_norm.drop(target, axis = 1)),
    "SIN_outliers_scal": (X_train_sel_SIN_outliers_scal.drop(target, axis = 1), X_test_sel_SIN_outliers_scal.drop(target, axis = 1))
}

Resultados_R = {}

for datasett, (X_train, X_test) in datasets_.items():
    model = Ridge(alpha = 0.1, max_iter = 300)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    Resultados_R[datasett] = {
        "Model" : model,
        "R2": r2_score(y_test, y_pred),
        "MAE": mean_absolute_error(y_test, y_pred),
        "RMSE": np.sqrt(mean_squared_error(y_test, y_pred))
    }

mejor_dataset_R = max(Resultados_R, key=lambda x: Resultados_R[x]["R2"])
mejor_r2_R = Resultados_R[mejor_dataset_R]["R2"]
mejor_modelo_R = Resultados_R[mejor_dataset_R]["Model"]

print("Resultados por dataset:")
for k, v in Resultados_R.items():
    print(f"{k:25s} -> R2: {v['R2']:.4f} | MAE: {v['MAE']:.4f} | RMSE: {v['RMSE']:.4f}")

print("\nMejor dataset:", mejor_dataset_R)
print("Mejor R2:", mejor_r2_R)

print("\nParámetros del mejor modelo:")
print("Intercepto (a):", mejor_modelo_R.intercept_)
print("Coeficientes (b):",mejor_modelo_R.coef_)
```
